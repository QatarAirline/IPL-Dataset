```python
def handle_ID(value):
    if(value== 'NEW' or value=='_'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
ab['ID']=ab['ID'].apply(handle_ID)
ab['ID'].head()
```

```python
plt.subplots(figsize=(20,8))
sns.countplot(x = 'ID', data = ab, palette='rainbow')
plt.title('ID numbers in list')
```
